# Implementacao do cga (Algoritmo genetico compacto)

```python
import numpy as np
```

```python
def cga(l,n,c):
    
    # 1-Inicializando o vetor de probabilidade
    l_prob = [0.5]*l
    convergiu = False
    ciclo = 0
    while convergiu == False and ciclo < c:
        
        # 2-Gerando 2 individuos a partir do vetor de probabilidades
        #print(l_prob)
        a = [np.random.choice([0,1], 1, p = [1-prob,prob])[0] for prob in l_prob]
        b = [np.random.choice([0,1], 1, p = [1-prob,prob])[0] for prob in l_prob]
        
        # 3-Competindo os dois individuos
        if (a.count(1) >= b.count(1)):
            ganhador = a
            perdedor = b
        else:
            ganhador = b
            perdedor = a
                
        # 4-Atualizando o vetor de probabilidades atrás do melhor individuo
        for i in range(l):
            if (ganhador[i] != perdedor[i]):
                if (ganhador[i] == 1):
                    l_prob[i] = round(l_prob[i] + 1/n,3)
                    if l_prob[i] > 1:
                        l_prob[i] = 1
                else:
                    l_prob[i] = round(l_prob[i] - 1/n,3)
                    if l_prob[i] < 0:
                        l_prob[i] = 0
        # 5-Checando se o vetor de probabilidades convergiu
        convergiu = True
        for i in range(l):
            if l_prob[i] > 0 and l_prob[i] < 1:
                convergiu = False
                break
        ciclo = ciclo + 1
        #print("Geração: ", ciclo, "a", a, "b", b, "p", l_prob, 1/n)
    #print(l_prob.count(1.0), ciclo)
    return(l_prob, ciclo)
```

## Resultados

```python
import time
# (4,8,...,100)
pop_size = np.arange(4, 101, 2).tolist()

l = 100
list_result = []
c = 10000

rodadas = 30

file = open('cga-onemax.txt','w')
for pop in pop_size:
    for _ in range(rodadas):
        start_time = time.time()
        l_prob, ciclo = cga(l,pop,c)
        duracao = time.time() - start_time
        list_result.append([pop,l_prob.count(1), ciclo, duracao])
        file.write('%s,%s,%s,%s,%s\n' %(pop,l,l_prob.count(1), ciclo, duracao))
        print("pop", pop, "n_bits",l_prob.count(1), "n_a", ciclo, "duração",duracao)
file.close()
    
```
